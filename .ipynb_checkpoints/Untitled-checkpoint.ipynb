```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
imgs = 'https://c8.alamy.com/comp/H69CFM/tianjinchina-march-202016-city-traffic-traffic-jams-a-stream-of-cars-H69CFM.jpg'
```

```python
results = model(imgs)
results.print
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()           
```

```python
plt.imshow(np.squeeze(results.render()))
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame=cap.read()
    results=model(frame)
    cv2.imshow('Yolo', np.squeeze(results.render()))
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
